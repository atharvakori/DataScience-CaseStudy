# Visualize the data using
# Histograms, Box Plots, and Scatter Plots to identify Patterns and Trends.

```python
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = pd.to_numeric(data[column].str.replace(',', ''), errors='coerce')

# Histograms
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
sns.histplot(data['Total Cases'].dropna(), bins=20, kde=True, color='blue')
plt.title('Total Cases Histogram')

plt.subplot(2, 2, 2)
sns.histplot(data['Total Deaths'].dropna(), bins=20, kde=True, color='red')
plt.title('Total Deaths Histogram')

plt.subplot(2, 2, 3)
sns.histplot(data['Total Recovered'].dropna(), bins=20, kde=True, color='green')
plt.title('Total Recovered Histogram')

plt.subplot(2, 2, 4)
sns.histplot(data['Total Tests'].dropna(), bins=20, kde=True, color='orange')
plt.title('Total Tests Histogram')

plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(15, 6))
plt.subplot(1, 3, 1)
sns.boxplot(data=data[['Total Cases', 'Total Deaths', 'Total Recovered']])
plt.title('Box Plot of Total Cases, Deaths, and Recovered')

plt.subplot(1, 3, 2)
sns.boxplot(data=data[['New Cases', 'New Deaths', 'New Recovered']])
plt.title('Box Plot of New Cases, Deaths, and Recovered')

plt.subplot(1, 3, 3)
sns.boxplot(data=data[['Total Tests']])
plt.title('Box Plot of Total Tests')

plt.tight_layout()
plt.show()

# Scatter plots
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='Total Cases', y='Total Deaths', data=data)
plt.title('Total Cases vs Total Deaths')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Total Cases', y='Total Recovered', data=data)
plt.title('Total Cases vs Total Recovered')

plt.tight_layout()
plt.show()
```

# Correlation Matrix

```python
data = data.dropna(subset=['Total Cases', 'Total Deaths', 'Total Recovered'])

total_cases = data['Total Cases']
total_deaths = data['Total Deaths']
total_recovered = data['Total Recovered']

# Calculate the correlation matrix
correlation_matrix = data[['Total Cases', 'Total Deaths', 'Total Recovered']].corr()

# Print the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Extract the correlation coefficient between Total Cases, Total Deaths, and Total Recovered
correlation_coefficient_cases_deaths = correlation_matrix.loc['Total Cases', 'Total Deaths']
correlation_coefficient_cases_recovered = correlation_matrix.loc['Total Cases', 'Total Recovered']
correlation_coefficient_deaths_recovered = correlation_matrix.loc['Total Deaths', 'Total Recovered']

# Print the correlation coefficients
print("\nCorrelation Coefficient (Total Cases, Total Deaths):", correlation_coefficient_cases_deaths)
print("Correlation Coefficient (Total Cases, Total Recovered):", correlation_coefficient_cases_recovered)
print("Correlation Coefficient (Total Deaths, Total Recovered):", correlation_coefficient_deaths_recovered)
```

# Correlation Matrix (Using HeadMap)

```python
columns_of_interest = ['Total Cases', 'Total Deaths', 'Total Recovered', 'Total Tests']

# Filter the DataFrame to include only the selected columns
selected_data = data[columns_of_interest]

# Compute the correlation matrix
correlation_matrix = selected_data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of COVID-19 Variables')
plt.show()
```
